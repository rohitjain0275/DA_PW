# **What is PANDAS?**

1. Most popular Python library for data analysis
2. Pandas is a Python library used for working with data sets.
3. It has functions for 
      - Analyzing, 
      - Cleaning, 
      - Exploring, 
      - Manipulating data..
4. Pandas is built on top of the Numpy package, means Numpy is required for operating the Pandas.

# **What's Pandas for??**
1. Pandas allows us to analyze big data and make conclusions based on statistical theories.
2. Calculate statistics and answer questions about the data, like -
    - What's the average, median, max, or min of each column?
    - Does column A correlate with column B?
    - What does the distribution of data in column C look like?
3. Clean the data by doing things like removing missing values and filtering rows or columns by some criteria
4. Visualize the data with help from Matplotlib. Plot bars, lines, histograms, bubbles, and more.
5. Store the cleaned, transformed data back into a CSV, other file or database

6. Pandas can clean messy data sets, and make them readable and relevant.
